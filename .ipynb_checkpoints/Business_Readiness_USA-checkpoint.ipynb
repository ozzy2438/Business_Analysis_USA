```python
# Step 1: Load & standardize CBP (2020-2023)
print("=== STEP 1: LOADING & STANDARDIZING CBP DATA ===")
print()

# Function to load and standardize CBP data
def load_and_standardize_cbp(filepath, year):
    """Load CBP file with proper dtypes and standardization"""
    df = pd.read_csv(filepath, 
                     dtype={'fipstate': str, 'fipscty': str, 'naics': str}, 
                     na_values=['N'])
    
    # Create 5-digit FIPS code
    df['FIPS'] = df['fipstate'].str.zfill(2) + df['fipscty'].str.zfill(3)
    
    # Add year column
    df['year'] = year
    
    # Numeric coercion for est and emp (already done by pd.read_csv with na_values=['N'])
    # Keep only required columns
    df_clean = df[['FIPS', 'naics', 'est', 'emp', 'year']].copy()
    
    print(f"Year {year}: Loaded {len(df_clean):,} records")
    print(f"  - Unique counties: {df_clean['FIPS'].nunique():,}")
    print(f"  - Missing est: {df_clean['est'].isna().sum():,} ({df_clean['est'].isna().mean()*100:.1f}%)")
    print(f"  - Missing emp: {df_clean['emp'].isna().sum():,} ({df_clean['emp'].isna().mean()*100:.1f}%)")
    
    return df_clean

# Load all CBP files
cbp_files = [
    ('/Users/osmanorka/Desktop/is-bazli-yeni-proje/cbp20co.csv', 2020),
    ('/Users/osmanorka/Desktop/is-bazli-yeni-proje/cbp21co.csv', 2021),
    ('/Users/osmanorka/Desktop/is-bazli-yeni-proje/cbp22co.csv', 2022),
    ('/Users/osmanorka/Desktop/is-bazli-yeni-proje/cbp23co.csv', 2023)
]

cbp_standardized = []
for filepath, year in cbp_files:
    df_std = load_and_standardize_cbp(filepath, year)
    cbp_standardized.append(df_std)
    print()

print(f"Total standardized records across all years: {sum(len(df) for df in cbp_standardized):,}") 
```

```python
# Step 2: County-level totals per year
print("=== STEP 2: COUNTY-LEVEL TOTALS PER YEAR ===")
print()

# Function to create county-level totals
def create_county_totals(df_list):
    """Sum establishments and employment across all NAICS per county-year"""
    county_totals = []
    
    for df in df_list:
        year = df['year'].iloc[0]
        print(f"Processing year {year}...")
        
        # Group by FIPS and sum est and emp across all NAICS
        county_year = df.groupby('FIPS')[['est', 'emp']].sum().reset_index()
        county_year['year'] = year
        
        # Rename columns to match specification
        county_year = county_year.rename(columns={'est': 'establishments', 'emp': 'employment'})
        
        # Reorder columns
        county_year = county_year[['FIPS', 'year', 'establishments', 'employment']]
        
        print(f"  - Counties: {len(county_year):,}")
        print(f"  - Missing establishments: {county_year['establishments'].isna().sum():,} ({county_year['establishments'].isna().mean()*100:.1f}%)")
        print(f"  - Missing employment: {county_year['employment'].isna().sum():,} ({county_year['employment'].isna().mean()*100:.1f}%)")
        print(f"  - Sample data:")
        print(county_year.head())
        print()
        
        county_totals.append(county_year)
    
    return county_totals

# Create county totals for each year
county_totals_list = create_county_totals(cbp_standardized)

print(f"Summary across all years:")
for i, df in enumerate(county_totals_list):
    year = df['year'].iloc[0]
    print(f"  {year}: {len(df):,} counties") 
```

```python
# Step 3: Build a 2020-2023 panel & growth
print("=== STEP 3: BUILDING 2020-2023 PANEL & GROWTH ===")
print()

# Concatenate all years into a single panel
county_year_panel = pd.concat(county_totals_list, ignore_index=True)
print(f"Total county-year observations: {len(county_year_panel):,}")
print(f"Years covered: {sorted(county_year_panel['year'].unique())}")
print(f"Unique counties across all years: {county_year_panel['FIPS'].nunique():,}")
print()

# Check counties that appear in both 2020 and 2023
counties_2020 = set(county_year_panel[county_year_panel['year'] == 2020]['FIPS'])
counties_2023 = set(county_year_panel[county_year_panel['year'] == 2023]['FIPS'])
counties_both = counties_2020.intersection(counties_2023)

print(f"Counties in 2020: {len(counties_2020):,}")
print(f"Counties in 2023: {len(counties_2023):,}")
print(f"Counties in both 2020 and 2023: {len(counties_both):,}")
print(f"Counties only in 2020: {len(counties_2020 - counties_2023):,}")
print(f"Counties only in 2023: {len(counties_2023 - counties_2020):,}")
print()

# Create growth analysis for counties that appear in both 2020 and 2023
print("Creating growth analysis for counties present in both 2020 and 2023...")

# Get 2020 and 2023 data for counties that appear in both years
data_2020 = county_year_panel[(county_year_panel['year'] == 2020) & 
                              (county_year_panel['FIPS'].isin(counties_both))].copy()
data_2023 = county_year_panel[(county_year_panel['year'] == 2023) & 
                              (county_year_panel['FIPS'].isin(counties_both))].copy()

print(f"2020 data for growth analysis: {len(data_2020):,} counties")
print(f"2023 data for growth analysis: {len(data_2023):,} counties")

# Merge 2020 and 2023 data
county_growth_2020_2023 = data_2020[['FIPS', 'establishments', 'employment']].merge(
    data_2023[['FIPS', 'establishments', 'employment']], 
    on='FIPS', 
    suffixes=('_2020', '_2023')
)

# Rename columns to match specification
county_growth_2020_2023 = county_growth_2020_2023.rename(columns={
    'establishments_2020': 'est_2020',
    'establishments_2023': 'est_2023',
    'employment_2020': 'emp_2020',
    'employment_2023': 'emp_2023'
})

print(f"Counties in growth analysis: {len(county_growth_2020_2023):,}")
print(f"Sample of merged data:")
print(county_growth_2020_2023.head())
print() 
```

```python
import numpy as np

# Calculate growth rates
print("Calculating growth rates...")

# Calculate establishment growth percentage
county_growth_2020_2023['estab_growth_pct'] = (
    county_growth_2020_2023['est_2023'] / county_growth_2020_2023['est_2020'] - 1
)

# Calculate employment growth percentage
county_growth_2020_2023['empl_growth_pct'] = (
    county_growth_2020_2023['emp_2023'] / county_growth_2020_2023['emp_2020'] - 1
)

print(f"Growth rates calculated for {len(county_growth_2020_2023):,} counties")
print(f"Sample of growth data:")
print(county_growth_2020_2023[['FIPS', 'est_2020', 'est_2023', 'estab_growth_pct', 
                               'emp_2020', 'emp_2023', 'empl_growth_pct']].head())
print()

# Check for any infinite or missing values
print("Data quality checks:")
print(f"Counties with infinite establishment growth: {np.isinf(county_growth_2020_2023['estab_growth_pct']).sum()}")
print(f"Counties with infinite employment growth: {np.isinf(county_growth_2020_2023['empl_growth_pct']).sum()}")
print(f"Counties with missing establishment growth: {county_growth_2020_2023['estab_growth_pct'].isna().sum()}")
print(f"Counties with missing employment growth: {county_growth_2020_2023['empl_growth_pct'].isna().sum()}")
print()

# Summary statistics
print("Growth rate summary statistics:")
print(f"Establishment growth - Mean: {county_growth_2020_2023['estab_growth_pct'].mean():.3f}, "
      f"Median: {county_growth_2020_2023['estab_growth_pct'].median():.3f}")
print(f"Employment growth - Mean: {county_growth_2020_2023['empl_growth_pct'].mean():.3f}, "
      f"Median: {county_growth_2020_2023['empl_growth_pct'].median():.3f}") 
```

```python
# Step 4: Prepare 2022 population & merge
print("=== STEP 4: PREPARING 2022 POPULATION & MERGE ===")
print()

# Create 5-digit FIPS from STATE+COUNTY (zero-padded)
pop_2022_clean = pop_2022.copy()
pop_2022_clean['FIPS'] = pop_2022_clean['STATE'].str.zfill(2) + pop_2022_clean['COUNTY'].str.zfill(3)

# Keep FIPS and TOT_POP, rename to pop_2022
pop_2022_final = pop_2022_clean[['FIPS', 'TOT_POP']].rename(columns={'TOT_POP': 'pop_2022'})

print(f"Population data prepared: {len(pop_2022_final):,} counties")
print(f"Sample of population data:")
print(pop_2022_final.head())
print()

# Left-join to the growth panel
print("Merging population data with growth analysis...")
county_growth_2020_2023_with_pop = county_growth_2020_2023.merge(
    pop_2022_final, 
    on='FIPS', 
    how='left'
)

print(f"Counties in growth analysis before merge: {len(county_growth_2020_2023):,}")
print(f"Counties in growth analysis after merge: {len(county_growth_2020_2023_with_pop):,}")
print(f"Counties with population data: {county_growth_2020_2023_with_pop['pop_2022'].notna().sum():,}")
print(f"Counties missing population data: {county_growth_2020_2023_with_pop['pop_2022'].isna().sum():,}")
print(f"Match rate: {county_growth_2020_2023_with_pop['pop_2022'].notna().mean()*100:.1f}%")
print()

print(f"Sample of merged data:")
print(county_growth_2020_2023_with_pop[['FIPS', 'est_2020', 'est_2023', 'estab_growth_pct', 'pop_2022']].head()) 
```

```python
# Step 5: Per-capita metrics
print("=== STEP 5: PER-CAPITA METRICS ===")
print()

# Calculate per-capita metrics
county_growth_2020_2023_with_pop['estab_per_1k'] = (
    county_growth_2020_2023_with_pop['est_2023'] / county_growth_2020_2023_with_pop['pop_2022'] * 1000
)

county_growth_2020_2023_with_pop['empl_per_1k'] = (
    county_growth_2020_2023_with_pop['emp_2023'] / county_growth_2020_2023_with_pop['pop_2022'] * 1000
)

print(f"Per-capita metrics calculated for {len(county_growth_2020_2023_with_pop):,} counties")
print(f"Sample of per-capita data:")
print(county_growth_2020_2023_with_pop[['FIPS', 'est_2023', 'emp_2023', 'pop_2022', 'estab_per_1k', 'empl_per_1k']].head())
print()

# Add simple data quality flags
print("Adding data quality flags...")
county_growth_2020_2023_with_pop['missing_pop'] = county_growth_2020_2023_with_pop['pop_2022'].isna()

print(f"Data quality flags:")
print(f"  - Counties with missing population: {county_growth_2020_2023_with_pop['missing_pop'].sum():,}")
print(f"  - Counties with valid per-capita metrics: {county_growth_2020_2023_with_pop['estab_per_1k'].notna().sum():,}")
print()

# Summary statistics for per-capita metrics
print("Per-capita metrics summary statistics:")
valid_data = county_growth_2020_2023_with_pop[county_growth_2020_2023_with_pop['pop_2022'].notna()]
print(f"Establishments per 1k - Mean: {valid_data['estab_per_1k'].mean():.1f}, Median: {valid_data['estab_per_1k'].median():.1f}")
print(f"Employment per 1k - Mean: {valid_data['empl_per_1k'].mean():.1f}, Median: {valid_data['empl_per_1k'].median():.1f}")
print(f"Range - Estab per 1k: {valid_data['estab_per_1k'].min():.1f} to {valid_data['estab_per_1k'].max():.1f}")
print(f"Range - Empl per 1k: {valid_data['empl_per_1k'].min():.1f} to {valid_data['empl_per_1k'].max():.1f}") 
```

```python
# Step 6: Clean outputs & save (no plots yet)
print("=== STEP 6: CLEAN OUTPUTS & FINAL RESULTS ===")
print()

# Assign the final dataframes with proper column names
county_growth_2020_2023_final = county_growth_2020_2023_with_pop.copy()

# Ensure column order matches specification
final_columns = ['FIPS', 'est_2020', 'est_2023', 'emp_2020', 'emp_2023', 
                'estab_growth_pct', 'empl_growth_pct', 'pop_2022', 
                'estab_per_1k', 'empl_per_1k']
county_growth_2020_2023_final = county_growth_2020_2023_final[final_columns]

print(f"Final datasets created:")
print(f"1. county_year_panel (long): {len(county_year_panel):,} rows, {len(county_year_panel.columns)} columns")
print(f"   Columns: {list(county_year_panel.columns)}")
print(f"2. county_growth_2020_2023_final (wide): {len(county_growth_2020_2023_final):,} rows, {len(county_growth_2020_2023_final.columns)} columns")
print(f"   Columns: {list(county_growth_2020_2023_final.columns)}")
print()

# Display top/bottom 10 counties by different metrics
print("=== TOP/BOTTOM 10 COUNTIES BY KEY METRICS ===")
print()

# Top/Bottom 10 by establishment growth
print("TOP 10 COUNTIES BY ESTABLISHMENT GROWTH (2020-2023):")
top_estab_growth = county_growth_2020_2023_final.nlargest(10, 'estab_growth_pct')
print(top_estab_growth[['FIPS', 'est_2020', 'est_2023', 'estab_growth_pct']].round(3))
print()

print("BOTTOM 10 COUNTIES BY ESTABLISHMENT GROWTH (2020-2023):")
bottom_estab_growth = county_growth_2020_2023_final.nsmallest(10, 'estab_growth_pct')
print(bottom_estab_growth[['FIPS', 'est_2020', 'est_2023', 'estab_growth_pct']].round(3))
print() 
```

```python
# Task 1: Build county  NAICS (2-digit) matrix (2023, establishments)
print("=== TASK 1: COUNTY  NAICS (2-DIGIT) MATRIX ===")
print()

# Load 2023 CBP data with proper handling
cbp_2023 = pd.read_csv('/Users/osmanorka/Desktop/is-bazli-yeni-proje/cbp23co.csv', 
                       dtype={'fipstate': str, 'fipscty': str, 'naics': str}, 
                       na_values=['N'])

# Create 5-digit FIPS
cbp_2023['FIPS'] = cbp_2023['fipstate'].str.zfill(2) + cbp_2023['fipscty'].str.zfill(3)

print(f"Loaded 2023 CBP data: {len(cbp_2023):,} records")
print(f"Sample of NAICS codes: {cbp_2023['naics'].head(10).tolist()}")
print()

# Extract 2-digit NAICS codes (exclude total codes like '------')
cbp_2023['naics2'] = cbp_2023['naics'].str[:2]

# Filter out total codes and keep only valid 2-digit NAICS
valid_naics2 = cbp_2023[
    (cbp_2023['naics2'].str.isdigit()) &  # Only numeric 2-digit codes
    (~cbp_2023['naics'].str.contains('---', na=False)) &  # Exclude total codes
    (cbp_2023['est'].notna())  # Valid establishment counts
].copy()

print(f"Records with valid 2-digit NAICS: {len(valid_naics2):,}")
print(f"Unique 2-digit NAICS codes: {sorted(valid_naics2['naics2'].unique())}")
print(f"Unique counties: {valid_naics2['FIPS'].nunique():,}")
print()

# Aggregate to county  naics2 level
county_naics2_2023 = valid_naics2.groupby(['FIPS', 'naics2'])['est'].sum().reset_index()
county_naics2_2023.columns = ['FIPS', 'naics2', 'est_2023']

print(f"County  NAICS2 combinations: {len(county_naics2_2023):,}")
print(f"Sample of county-naics2 data:")
print(county_naics2_2023.head(10))
print() 
```

```python
# Create county shares and pivot table
print("Creating county shares and pivot table...")
print()

# Calculate total establishments per county
county_totals_2023 = county_naics2_2023.groupby('FIPS')['est_2023'].sum().reset_index()
county_totals_2023.columns = ['FIPS', 'total_est_2023']

print(f"Counties with establishment data: {len(county_totals_2023):,}")
print(f"Sample of county totals:")
print(county_totals_2023.head())
print()

# Merge back to get shares
county_naics2_with_shares = county_naics2_2023.merge(county_totals_2023, on='FIPS')
county_naics2_with_shares['share_2023'] = county_naics2_with_shares['est_2023'] / county_naics2_with_shares['total_est_2023']

# Keep tidy long version
county_industry_mix_long = county_naics2_with_shares[['FIPS', 'naics2', 'est_2023', 'share_2023']].copy()

print(f"Long format industry mix: {len(county_industry_mix_long):,} records")
print(f"Sample of long format data:")
print(county_industry_mix_long.head(10))
print()

# Create pivot table (wide format)
county_naics_matrix = county_naics2_2023.pivot(index='FIPS', columns='naics2', values='est_2023').fillna(0)
print(f"Pivot matrix shape: {county_naics_matrix.shape}")
print(f"NAICS2 codes in matrix: {sorted(county_naics_matrix.columns.tolist())}")
print(f"Sample of pivot matrix:")
print(county_naics_matrix.head())
print()

# Create shares matrix (row-normalized)
county_shares_matrix = county_naics_matrix.div(county_naics_matrix.sum(axis=1), axis=0)
print(f"Shares matrix shape: {county_shares_matrix.shape}")
print(f"Sample of shares matrix (should sum to 1.0 per row):")
print(county_shares_matrix.head())
print(f"Row sums check (first 5): {county_shares_matrix.head().sum(axis=1).tolist()}") 
```

```python
# Task 2: US benchmark (2023)
print("=== TASK 2: US BENCHMARK (2023) ===")
print()

# Compute US total establishments by naics2 (sum across all counties)
us_totals_by_naics2 = county_naics2_2023.groupby('naics2')['est_2023'].sum().reset_index()
us_totals_by_naics2.columns = ['naics2', 'us_est_2023']

# Calculate total US establishments
total_us_est = us_totals_by_naics2['us_est_2023'].sum()

# Derive US_share_2023 for each naics2
us_totals_by_naics2['US_share_2023'] = us_totals_by_naics2['us_est_2023'] / total_us_est

print(f"Total US establishments (2023): {total_us_est:,}")
print(f"US benchmark by NAICS2:")
print(us_totals_by_naics2.sort_values('US_share_2023', ascending=False))
print()

# Verify shares sum to 1.0
print(f"US shares sum check: {us_totals_by_naics2['US_share_2023'].sum():.6f}")
print()

# Show top 5 industries by share
print("Top 5 industries by US share:")
top_5_industries = us_totals_by_naics2.nlargest(5, 'US_share_2023')
for _, row in top_5_industries.iterrows():
    print(f"  NAICS {row['naics2']}: {row['US_share_2023']:.1%} ({row['us_est_2023']:,} establishments)") 
```

```python
# Task 3: Location Quotients (LQ)
print("=== TASK 3: LOCATION QUOTIENTS (LQ) ===")
print()

# Merge county shares with US benchmark shares
county_industry_mix_with_lq = county_industry_mix_long.merge(
    us_totals_by_naics2[['naics2', 'US_share_2023']], 
    on='naics2'
)

# Calculate Location Quotients
county_industry_mix_with_lq['LQ_2023'] = county_industry_mix_with_lq['share_2023'] / county_industry_mix_with_lq['US_share_2023']

print(f"Location Quotients calculated for {len(county_industry_mix_with_lq):,} county-industry combinations")
print(f"Sample of LQ data:")
print(county_industry_mix_with_lq.head(10))
print()

# Check for extreme LQs
print("LQ distribution analysis:")
print(f"Min LQ: {county_industry_mix_with_lq['LQ_2023'].min():.2f}")
print(f"Max LQ: {county_industry_mix_with_lq['LQ_2023'].max():.2f}")
print(f"Mean LQ: {county_industry_mix_with_lq['LQ_2023'].mean():.2f}")
print(f"Median LQ: {county_industry_mix_with_lq['LQ_2023'].median():.2f}")
print(f"LQs >= 10: {(county_industry_mix_with_lq['LQ_2023'] >= 10).sum():,}")
print(f"LQs >= 5: {(county_industry_mix_with_lq['LQ_2023'] >= 5).sum():,}")
print()

# Create capped LQs at 10 to avoid outlier distortion
county_industry_mix_with_lq['LQ_2023_capped'] = county_industry_mix_with_lq['LQ_2023'].clip(upper=10)

print(f"Capped LQs at 10.0:")
print(f"Max capped LQ: {county_industry_mix_with_lq['LQ_2023_capped'].max():.2f}")
print(f"Values capped: {(county_industry_mix_with_lq['LQ_2023'] > 10).sum():,}")
print()

# Show examples of high LQs
print("Top 10 highest Location Quotients (uncapped):")
top_lqs = county_industry_mix_with_lq.nlargest(10, 'LQ_2023')
print(top_lqs[['FIPS', 'naics2', 'share_2023', 'US_share_2023', 'LQ_2023']].round(3)) 
```

```python
# Task 4: Diversity (HHI)
print("=== TASK 4: DIVERSITY (HHI) ===")
print()

# Calculate HHI for each county using the shares matrix
# HHI = sum(share^2) for each county (row)
county_hhi = county_shares_matrix.apply(lambda row: (row ** 2).sum(), axis=1).reset_index()
county_hhi.columns = ['FIPS', 'HHI']

# Calculate diversity = 1 - HHI for "variety" interpretation
county_hhi['diversity'] = 1 - county_hhi['HHI']

print(f"Diversity metrics calculated for {len(county_hhi):,} counties")
print(f"Sample of diversity data:")
print(county_hhi.head(10))
print()

# Summary statistics for diversity metrics
print("Diversity metrics summary statistics:")
print(f"HHI - Mean: {county_hhi['HHI'].mean():.3f}, Median: {county_hhi['HHI'].median():.3f}")
print(f"HHI - Min: {county_hhi['HHI'].min():.3f}, Max: {county_hhi['HHI'].max():.3f}")
print(f"Diversity - Mean: {county_hhi['diversity'].mean():.3f}, Median: {county_hhi['diversity'].median():.3f}")
print(f"Diversity - Min: {county_hhi['diversity'].min():.3f}, Max: {county_hhi['diversity'].max():.3f}")
print()

# Show most and least diverse counties
print("Top 10 most diverse counties (highest diversity = lowest HHI):")
most_diverse = county_hhi.nlargest(10, 'diversity')
print(most_diverse[['FIPS', 'HHI', 'diversity']].round(3))
print()

print("Top 10 least diverse counties (lowest diversity = highest HHI):")
least_diverse = county_hhi.nsmallest(10, 'diversity')
print(least_diverse[['FIPS', 'HHI', 'diversity']].round(3)) 
```

```python
# Task 5: Strategy buckets (binary flags per county)
print("=== TASK 5: STRATEGY BUCKETS (BINARY FLAGS) ===")
print()

# Define strategy buckets with their NAICS codes
strategy_buckets = {
    'Retail_Services': ['44', '45', '72', '81', '53'],
    'Pro_Tech': ['51', '52', '54'],
    'Health_Edu': ['61', '62'],
    'Manufacturing': ['31', '32', '33'],
    'Logistics_Trade': ['48', '49', '42']
}

print("Strategy bucket definitions:")
for bucket, naics_codes in strategy_buckets.items():
    print(f"  {bucket}: NAICS {naics_codes}")
print()

# Create a dataframe to store bucket results
bucket_results = []

# Process each county
for fips in county_shares_matrix.index:
    county_data = {'FIPS': fips}
    
    # Get LQ data for this county
    county_lqs = county_industry_mix_with_lq[county_industry_mix_with_lq['FIPS'] == fips]
    
    # Process each strategy bucket
    for bucket_name, naics_codes in strategy_buckets.items():
        # Get LQs for NAICS codes in this bucket
        bucket_lqs = county_lqs[county_lqs['naics2'].isin(naics_codes)]
        
        # Check if any LQ >= 1.25 (specialized)
        specialized = (bucket_lqs['LQ_2023'] >= 1.25).any() if len(bucket_lqs) > 0 else False
        county_data[f'{bucket_name}_specialized'] = 1 if specialized else 0
        
        # Calculate mean share across NAICS codes in this bucket
        bucket_shares = county_lqs[county_lqs['naics2'].isin(naics_codes)]['share_2023']
        mean_share = bucket_shares.mean() if len(bucket_shares) > 0 else 0.0
        county_data[f'{bucket_name}_mean_share'] = mean_share
    
    bucket_results.append(county_data)

# Convert to DataFrame
county_strategy_buckets = pd.DataFrame(bucket_results)

print(f"Strategy buckets calculated for {len(county_strategy_buckets):,} counties")
print(f"Sample of strategy bucket data:")
print(county_strategy_buckets.head())
print()

# Summary statistics for specializations
print("Specialization summary (counties with LQ >= 1.25):")
for bucket_name in strategy_buckets.keys():
    specialized_count = county_strategy_buckets[f'{bucket_name}_specialized'].sum()
    total_counties = len(county_strategy_buckets)
    pct = (specialized_count / total_counties) * 100
    print(f"  {bucket_name}: {specialized_count:,} counties ({pct:.1f}%)")
print()

# Show mean shares summary
print("Mean share summary across buckets:")
for bucket_name in strategy_buckets.keys():
    mean_col = f'{bucket_name}_mean_share'
    mean_val = county_strategy_buckets[mean_col].mean()
    median_val = county_strategy_buckets[mean_col].median()
    print(f"  {bucket_name}: Mean={mean_val:.3f}, Median={median_val:.3f}") 
```

```python
# Task 6: Clustering (economic types)
print("=== TASK 6: CLUSTERING (ECONOMIC TYPES) ===")
print()

# Import sklearn for k-means clustering
from sklearn.cluster import KMeans
import numpy as np

# Run k-means (k=5) on the row-normalized share matrix (no totals)
# Use a fixed random seed for reproducibility
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(county_shares_matrix)

# Create cluster results dataframe
cluster_results = pd.DataFrame({
    'FIPS': county_shares_matrix.index,
    'cluster_5': cluster_labels
})

print(f"K-means clustering completed for {len(cluster_results):,} counties")
print(f"Cluster distribution:")
cluster_counts = cluster_results['cluster_5'].value_counts().sort_index()
for cluster_id, count in cluster_counts.items():
    print(f"  Cluster {cluster_id}: {count:,} counties ({count/len(cluster_results)*100:.1f}%)")
print()

# Show 2 example FIPS per cluster
print("Example counties per cluster:")
for cluster_id in sorted(cluster_results['cluster_5'].unique()):
    cluster_counties = cluster_results[cluster_results['cluster_5'] == cluster_id]['FIPS'].head(2).tolist()
    print(f"  Cluster {cluster_id}: {cluster_counties}")
print()

# Analyze cluster characteristics by looking at mean industry shares
print("Cluster characteristics (mean industry shares):")
cluster_characteristics = []
for cluster_id in sorted(cluster_results['cluster_5'].unique()):
    cluster_mask = cluster_results['cluster_5'] == cluster_id
    cluster_fips = cluster_results[cluster_mask]['FIPS'].tolist()
    cluster_shares = county_shares_matrix.loc[cluster_fips].mean()
    
    # Get top 3 industries for this cluster
    top_industries = cluster_shares.nlargest(3)
    print(f"  Cluster {cluster_id} (n={cluster_mask.sum():,}):")
    for naics, share in top_industries.items():
        print(f"    NAICS {naics}: {share:.1%}")
    print()

print(f"Clustering results: {len(cluster_results):,} counties assigned to 5 clusters") 
```

```python
# Task 7: Merge features with county_growth_2020_2023_final
print("=== TASK 7: MERGING FEATURES WITH GROWTH DATA ===")
print()

# Merge diversity metrics
county_growth_with_diversity = county_growth_2020_2023_final.merge(
    county_hhi[['FIPS', 'diversity']], 
    on='FIPS', 
    how='left'
)

# Merge cluster results
county_growth_with_clusters = county_growth_with_diversity.merge(
    cluster_results[['FIPS', 'cluster_5']], 
    on='FIPS', 
    how='left'
)

# Merge strategy bucket flags and mean shares
county_growth_with_buckets = county_growth_with_clusters.merge(
    county_strategy_buckets, 
    on='FIPS', 
    how='left'
)

print(f"Merged features with growth data: {len(county_growth_with_buckets):,} counties")
print(f"Sample of merged data:")
print(county_growth_with_buckets[['FIPS', 'estab_growth_pct', 'diversity', 'cluster_5', 
                                  'Retail_Services_specialized', 'Manufacturing_specialized']].head())
print()

# Get top 5 LQs per county
print("Calculating top 5 LQs per county...")
top_lqs_per_county = []

for fips in county_growth_with_buckets['FIPS']:
    county_lq_data = county_industry_mix_with_lq[county_industry_mix_with_lq['FIPS'] == fips]
    if len(county_lq_data) > 0:
        top_5_lqs = county_lq_data.nlargest(5, 'LQ_2023')[['naics2', 'LQ_2023']]
        
        # Create top 5 LQ columns
        lq_dict = {'FIPS': fips}
        for i, (_, row) in enumerate(top_5_lqs.iterrows(), 1):
            lq_dict[f'top{i}_naics'] = row['naics2']
            lq_dict[f'top{i}_LQ'] = row['LQ_2023']
        
        # Fill remaining slots with None if less than 5 industries
        for i in range(len(top_5_lqs) + 1, 6):
            lq_dict[f'top{i}_naics'] = None
            lq_dict[f'top{i}_LQ'] = None
            
        top_lqs_per_county.append(lq_dict)
    else:
        # Handle counties with no LQ data
        lq_dict = {'FIPS': fips}
        for i in range(1, 6):
            lq_dict[f'top{i}_naics'] = None
            lq_dict[f'top{i}_LQ'] = None
        top_lqs_per_county.append(lq_dict)

top_lqs_df = pd.DataFrame(top_lqs_per_county)

print(f"Top 5 LQs calculated for {len(top_lqs_df):,} counties")
print(f"Sample of top LQs:")
print(top_lqs_df[['FIPS', 'top1_naics', 'top1_LQ', 'top2_naics', 'top2_LQ']].head())
print()

# Final merge with top LQs
county_growth_2020_2023_with_industry = county_growth_with_buckets.merge(
    top_lqs_df, 
    on='FIPS', 
    how='left'
)

print(f"Final consolidated table: {len(county_growth_2020_2023_with_industry):,} counties")
print(f"Total columns: {len(county_growth_2020_2023_with_industry.columns)}")
print(f"Column names: {list(county_growth_2020_2023_with_industry.columns)}") 
```

```python
# Step 2: Feature prep (null-safe)
print("=== STEP 2: FEATURE PREP (NULL-SAFE) ===")

# Start from previous output, work with county_master_with_names
df = county_master_with_names.copy()

# Keep/confirm required fields and safe versions
required_cols = [
    'FIPS', 'STATEFP', 'COUNTYFP', 'STNAME', 'CTYNAME',
    'estab_growth_pct', 'empl_growth_pct',
    'estab_per_1k', 'empl_per_1k', 'diversity', 'cluster_5',
    'Retail_Services_specialized', 'Pro_Tech_specialized', 'Health_Edu_specialized', 'Manufacturing_specialized', 'Logistics_Trade_specialized',
    'total_specializations',
    'top1_naics', 'top1_LQ'
]
for req in required_cols:
    if req not in df.columns:
        df[req] = None

# Create has_industry_data: must have diversity and at least one industry share (using diversity as proxy for industry richness)
df['has_industry_data'] = df['diversity'].notna() & (df['diversity'] > 0)

# Create top1_LQ_capped (if top1_LQ present)
df['top1_LQ_capped'] = df['top1_LQ'].clip(upper=10) if 'top1_LQ' in df else None

print("Retained/created columns:")
print(df[required_cols + ['has_industry_data', 'top1_LQ_capped']].head(10))
print()

# If has_industry_data==False, set diversity and multi_specialized (total_specializations >= 2) contributions to 0 (to be implemented in the scoring step)
 
```

```python
# Step 3: Scoring (null-safe, documented weights, clipped z-scores)
print("=== STEP 3: SCORING: BUSINESS EXPANSION READINESS ===\n")

score_cfg = {
    'estab_growth_pct': 0.35,
    'empl_growth_pct': 0.20,
    'estab_per_1k': 0.20,
    'diversity': 0.15,
    'multi_specialized': 0.10 # 1 if total_specializations >=2 else 0
}

def zscore(col):
    # Null-safe z-score & clipping to 3
    vals = df[col]
    m = vals.mean(skipna=True)
    s = vals.std(ddof=0, skipna=True)
    z = (vals - m) / s
    return z.clip(-3, 3)

# Multi-specialized flag
multi_spec = (df['total_specializations'] >= 2).astype(float)

# Null-safe score components
z_estab_growth = zscore('estab_growth_pct')
z_empl_growth = zscore('empl_growth_pct')
z_estab_per_1k = zscore('estab_per_1k')
z_div = zscore('diversity')

score_missing_industry = (~df['has_industry_data']).astype(int)

# If county lacks industry data, set diversity/multi-specialized to zero in score
z_div_adj = z_div.where(df['has_industry_data'], 0)
multi_spec_adj = multi_spec.where(df['has_industry_data'], 0)

# Final score as weighted sum:
df['opportunity_score'] = (
    score_cfg['estab_growth_pct'] * z_estab_growth +
    score_cfg['empl_growth_pct'] * z_empl_growth +
    score_cfg['estab_per_1k']   * z_estab_per_1k +
    score_cfg['diversity']      * z_div_adj +
    score_cfg['multi_specialized'] * multi_spec_adj
)

# Score for missing industry data flag
df['score_missing_industry'] = score_missing_industry

# Percentile (0-100, higher is better)
df['opportunity_pctile'] = df['opportunity_score'].rank(pct=True) * 100
# Tier assignment
bins = [0, 40, 70, 90, 100]
tiers = ['D', 'C', 'B', 'A']
df['opportunity_tier'] = pd.cut(df['opportunity_pctile'], bins=bins, labels=tiers, right=True, include_lowest=True)

print('Score summary:')
print(df[['opportunity_score', 'opportunity_pctile', 'opportunity_tier', 'score_missing_industry']].describe())
print(df[['FIPS', 'opportunity_score', 'opportunity_pctile', 'opportunity_tier', 'score_missing_industry']].head(10))
print()
print('Tier counts:')
print(df['opportunity_tier'].value_counts(dropna=False)) 
```

```python
# Step 4: Master table output
print("=== STEP 4: MASTER COUNTY BUSINESS READINESS TABLE ===\n")

# Clean final columns (order as requested)
final_cols = [
    'FIPS', 'STATEFP', 'COUNTYFP', 'STNAME', 'CTYNAME',
    'estab_growth_pct', 'empl_growth_pct', 'estab_per_1k', 'empl_per_1k',
    'diversity', 'cluster_5',
    'Retail_Services_specialized','Pro_Tech_specialized','Health_Edu_specialized',
    'Manufacturing_specialized','Logistics_Trade_specialized','total_specializations',
    'top1_naics','top1_LQ_capped','has_industry_data','score_missing_industry',
    'opportunity_score','opportunity_pctile','opportunity_tier'
]

# Some counties may lack some fields - ensure columns exist in df
for col in final_cols:
    if col not in df.columns:
        df[col] = None

county_business_readiness_2023 = df[final_cols].copy()
county_business_readiness_2023.to_csv('county_business_readiness_2023.csv', index=False)

print(f"Saved county_business_readiness_2023.csv! Shape: {county_business_readiness_2023.shape}")
print(county_business_readiness_2023.head(10))
print('\nColumns:')
print(list(county_business_readiness_2023.columns))
print("\nTier distribution:")
print(county_business_readiness_2023['opportunity_tier'].value_counts(dropna=False)) 
```

```python
import os
import pandas as pd
import geopandas as gpd
from datetime import datetime

if not os.path.exists('exports'):
    os.makedirs('exports')

print('=== QA Step 1: Load Master and Shapefile ===')
# Load master table (metric columns to numeric, IDs to str)
metrics = ['opportunity_score','opportunity_pctile','estab_growth_pct','empl_growth_pct',
    'estab_per_1k','diversity','total_specializations','top1_LQ_capped']
df = pd.read_csv('county_business_readiness_2023.csv', dtype=str)
for col in metrics:
    if col in df: df[col] = pd.to_numeric(df[col], errors='coerce')
df['FIPS'] = df['FIPS'].astype(str).str.zfill(5)
print('Master loaded:', df.shape)
print('Null FIPS:', df['FIPS'].isna().sum())
print('Unique FIPS:', df['FIPS'].nunique(), 'Duplicates:', df.duplicated('FIPS').sum())
print(df.head(3))

# Load shapefile, ensure GEOID padded and CRS is EPSG:4326
shp_candidates = [f for f in os.listdir('.') if f.endswith('.shp')]
if os.path.isdir('cb_2022_us_county_5m'):
    shp_candidates += [os.path.join('cb_2022_us_county_5m', f) for f in os.listdir('cb_2022_us_county_5m') if f.endswith('.shp')]
assert shp_candidates, 'No shapefile found.'
shp_path = shp_candidates[0]
gdf = gpd.read_file(shp_path)
gdf['GEOID'] = gdf['GEOID'].astype(str).str.zfill(5)
if gdf.crs is None:
    gdf = gdf.set_crs('EPSG:4326')
elif gdf.crs != 'EPSG:4326':
    gdf = gdf.to_crs('EPSG:4326')
print('Shapefile loaded:', gdf.shape)
print('Unique GEOID:', gdf['GEOID'].nunique())
print(gdf[['GEOID', 'STATEFP', 'COUNTYFP','geometry']].head(3)) 
```

```python
print('=== QA Step 2: Join to Shapefile and Coverage Report ===')

# Join master table to shapefile on FIPS == GEOID
master_joined = pd.merge(df, gdf[['GEOID']], left_on='FIPS', right_on='GEOID', how='left')
matched = master_joined['GEOID'].notna().sum()
not_matched = (~master_joined['GEOID'].notna()).sum()
print(f'Matched counties (have geometry): {matched}')
print(f'Counties without geometry: {not_matched}')

# Write missing-geometry list
missing_geom = master_joined[master_joined['GEOID'].isna()][['FIPS', 'STNAME', 'CTYNAME']]
missing_geom.to_csv('exports/missing_geometry_counties.csv', index=False)
print(f"Missing geometry list written: {missing_geom.shape[0]} counties  'exports/missing_geometry_counties.csv'")
print(missing_geom.head())
 
```

```python
print('=== QA Step 3: LQ Outlier Check & Z-Score Inputs Summary ===')

# Outlier check: summarize cap effect for top1_LQ_capped
lq_col = 'top1_LQ_capped' if 'top1_LQ_capped' in df else 'top1_LQ'
capped = df[lq_col] == 10
not_capped = df[lq_col] < 10
print(f'LQ-capped at 10: {capped.sum()} counties; Uncapped: {not_capped.sum()} counties\n')
for col in ['opportunity_score','diversity','estab_growth_pct','estab_per_1k']:
    m1, s1 = df.loc[capped, col].mean(), df.loc[capped, col].std()
    m0, s0 = df.loc[not_capped, col].mean(), df.loc[not_capped, col].std()
    print(f"{col:20} capped: mean={m1:.3f}, std={s1:.3f} | uncapped: mean={m0:.3f}, std={s0:.3f}")
print()

# Z-score checks (mean/std, null-safe, 3 decimals)
def zscore_stats(col):
    vals = df[col].dropna()
    z = (vals - vals.mean()) / vals.std(ddof=0)
    print(f"z_{col:16} mean={z.mean():.3f}, std={z.std(ddof=0):.3f}")

for metric in ['estab_growth_pct','empl_growth_pct','estab_per_1k','diversity']:
    zscore_stats(metric) 
```

```python
print('=== QA Step 4: Curated Dashboard Geo Exports ===')
import geopandas as gpd
import os

dashboard_cols = [
    'FIPS','STNAME','CTYNAME','opportunity_score','opportunity_pctile','opportunity_tier',
    'estab_growth_pct','empl_growth_pct','estab_per_1k','diversity','total_specializations','top1_naics','top1_LQ_capped'
]

# Ensure numeric metrics are correct
df_geo = df.copy()
for f in ['opportunity_score','opportunity_pctile','estab_growth_pct','empl_growth_pct','estab_per_1k','diversity','total_specializations','top1_LQ_capped']:
    if f in df_geo:
        df_geo[f] = pd.to_numeric(df_geo[f], errors='coerce')

shp_candidates = [f for f in os.listdir('.') if f.endswith('.shp')]
if os.path.isdir('cb_2022_us_county_5m'):
    shp_candidates += [os.path.join('cb_2022_us_county_5m', f) for f in os.listdir('cb_2022_us_county_5m') if f.endswith('.shp')]
shp_path = shp_candidates[0]
gdf_county = gpd.read_file(shp_path)
gdf_county['GEOID'] = gdf_county['GEOID'].astype(str).str.zfill(5)
if gdf_county.crs is None:
    gdf_county = gdf_county.set_crs('EPSG:4326')
elif gdf_county.crs != 'EPSG:4326':
    gdf_county = gdf_county.to_crs('EPSG:4326')
# Join and drop missing geoms
geo_out = gdf_county.merge(df_geo[dashboard_cols], left_on='GEOID', right_on='FIPS', how='left')
geo_out = geo_out[geo_out['FIPS'].notna()].copy()
print(f"Geo features (dashboard export): {len(geo_out)} written to GeoJSON/Parquet")

geo_out.to_file('exports/geo_county_business_readiness_2023.geojson', driver='GeoJSON')
geo_out.to_parquet('exports/geo_county_business_readiness_2023.parquet', index=False)
print(geo_out[dashboard_cols].head()) 
```

```python
import os
import pandas as pd
from sqlalchemy import create_engine, VARCHAR, text
import pyodbc
import getpass

server = 'localhost'
database = 'Business_Readiness_USA'
username = input('Enter SQL Server username: ')
password = getpass.getpass('Enter SQL Server password: ')
conn_str = f"mssql+pyodbc://{username}:{password}@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server"
engine = create_engine(conn_str, fast_executemany=True)

csvs = [fn for fn in os.listdir('.') if fn.endswith('.csv') and not fn.startswith('missing_geom')]
print('CSV files found:', csvs)

for csv in csvs:
    table = csv.replace('.csv', '')
    path = csv
    print(f'\nUploading {csv} to table {table}...')
    df = pd.read_csv(path, dtype=str)
    for col in df.columns:
        if col.lower().startswith('fips') or col.lower().endswith('id'):
            df[col] = df[col].astype(str).str.zfill(5)
        else:
            try:
                df[col] = pd.to_numeric(df[col])
            except:
                pass
    dtypes = {col:VARCHAR(5) for col in df.columns if col.lower() == 'fips'}
    df.to_sql(table, engine, if_exists='replace', index=False, dtype=dtypes)
    # FIX: Use engine.connect() and connection.execute() instead of engine.execute()
    with engine.connect() as conn:
        res = conn.execute(text(f'SELECT COUNT(*) FROM {table}'))
        row_ct = res.fetchone()[0]
    print(f'  Table {table}: {row_ct} rows uploaded. First 5 rows:')
    print(df.head(5))
```
